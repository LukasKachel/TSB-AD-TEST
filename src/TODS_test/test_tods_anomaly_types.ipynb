# TODS Anomaly Type Test

Tests one statistical method on TODS point-anomaly datasets vs. sequence-anomaly datasets.

```python
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
from IPython.display import display

from TSB_AD.evaluation.metrics import get_metrics
from TSB_AD.model_wrapper import run_Unsupervise_AD

warnings.filterwarnings("ignore")

# Select method to test: "Sub_PCA", "POLY", "KMeansAD", "KShapeAD"
METHOD_NAME = "KShapeAD"

DATA_DIR  = Path("../../datasets/TSB-AD-U")
POINT_IDS = [1, 3, 5, 6, 10]
SEQ_IDS   = [2, 4, 7, 8, 9, 11, 12, 13, 14, 15]

RESULTS_DIR = Path("../../results/TODS") / METHOD_NAME
RESULTS_DIR.mkdir(parents=True, exist_ok=True)
print(f"Results directory: {RESULTS_DIR.resolve()}")
```

## Point Anomaly Datasets

```python
rows = []
for tods_id in POINT_IDS:
    path = next(DATA_DIR.glob(f"*_TODS_id_{tods_id}_*.csv"))
    df = pd.read_csv(path).dropna()
    data   = df.iloc[:, :-1].values.astype(float)
    labels = df["Label"].astype(int).to_numpy()

    scores = run_Unsupervise_AD(METHOD_NAME, data)
    if isinstance(scores, str):
        print(f"Error on TODS_id_{tods_id}: {scores}")
        continue

    scores = np.asarray(scores).ravel()
    n = min(len(scores), len(labels))
    metrics = get_metrics(scores[:n], labels[:n])
    rows.append({"dataset": f"TODS_id_{tods_id}", **metrics})

df_point = pd.DataFrame(rows)
display(df_point)
print("\nMean:")
display(df_point.mean(numeric_only=True).to_frame().T)

point_output_path = RESULTS_DIR / "point_results.csv"
df_point.to_csv(point_output_path, index=False)
print(f"Saved point results to: {point_output_path.resolve()}")
```

## Sequence Anomaly Datasets

```python
rows = []
for tods_id in SEQ_IDS:
    path = next(DATA_DIR.glob(f"*_TODS_id_{tods_id}_*.csv"))
    df = pd.read_csv(path).dropna()
    data   = df.iloc[:, :-1].values.astype(float)
    labels = df["Label"].astype(int).to_numpy()

    scores = run_Unsupervise_AD(METHOD_NAME, data)
    if isinstance(scores, str):
        print(f"Error on TODS_id_{tods_id}: {scores}")
        continue

    scores = np.asarray(scores).ravel()
    n = min(len(scores), len(labels))
    metrics = get_metrics(scores[:n], labels[:n])
    rows.append({"dataset": f"TODS_id_{tods_id}", **metrics})

df_seq = pd.DataFrame(rows)
display(df_seq)
print("\nMean:")
display(df_seq.mean(numeric_only=True).to_frame().T)

seq_output_path = RESULTS_DIR / "sequence_results.csv"
df_seq.to_csv(seq_output_path, index=False)
print(f"Saved sequence results to: {seq_output_path.resolve()}")
```
